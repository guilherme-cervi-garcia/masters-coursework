- CCM real reavaliada, escala {0, 1, 2, 3}
- Vetor A escala 1-4, 4 LLMs pos-revisao
- 144 configuracoes admissiveis

## Front Pareto por cenario

```python
# Front Pareto por cenario
for sc in SCENARIOS:
    rows = []
    # Numero maximo de configs no front entre versoes
    max_len = max(len(d['pareto_front'][sc]) for d in VERSIONS.values())
    for rank in range(max_len):
        row = {'Rank': rank + 1}
        for v, d in VERSIONS.items():
            front = d['pareto_front'][sc]
            if rank < len(front):
                cfg = front[rank]
                row[d['label']] = f"{cfg[0]}, {cfg[1]}, {cfg[2]}, {cfg[3]}"
            else:
                row[d['label']] = '-'
        rows.append(row)
    df = pd.DataFrame(rows).set_index('Rank')
    print(f"\n{sc} -- {SC_NAMES[sc]}")
    print(f"Tamanho do front: "
          f"V1={len(VERSIONS['V1']['pareto_front'][sc])} | "
          f"V2={len(VERSIONS['V2']['pareto_front'][sc])} | "
          f"V3={len(VERSIONS['V3']['pareto_front'][sc])}")
    print(df.to_string())
```

## Achados da evolucao

```python
# Achados da evolucao -- derivados dos dados

def da_dominante(versions, modulo_idx, modulo_nome, sc='SC1'):
    """DA mais frequente no rank 1 do scoring_top5 por modulo."""
    result = {}
    for v, d in versions.items():
        contagem = {}
        for sc_id in SCENARIOS:
            da = d['scoring_top5'][sc_id][0][modulo_idx]
            contagem[da] = contagem.get(da, 0) + 1
        dominante = max(contagem, key=contagem.get)
        result[v] = (dominante, contagem[dominante])
    return result

modulos = [
    (0, 'D'), (1, 'I'), (2, 'ID'), (3, 'R')
]

print("Achados da evolucao")
print("=" * 70)

# Espaco de solucoes
print("\nEspaco de solucoes")
print("-" * 20)
for v, d in VERSIONS.items():
    reducao = 100*(1 - d['n_admissible']/d['n_total'])
    print(f"  {d['label']}: {d['n_admissible']} / {d['n_total']} "
          f"({reducao:.1f}% de reducao)")

# Estabilidade dos rankings
print("\nEstabilidade dos rankings (Spearman minimo)")
print("-" * 44)
for v, d in VERSIONS.items():
    print(f"  {d['label']}: {d['spearman_min']} "
          f"({d['spearman_min_pair'][0]} x {d['spearman_min_pair'][1]})")

# Modulos dominantes
print("\nDA dominante no rank 1 do scoring (por modulo)")
print("-" * 48)
for idx, mod in modulos:
    print(f"\n  Modulo {mod}:")
    dom = da_dominante(VERSIONS, idx, mod)
    for v, d in VERSIONS.items():
        da, n = dom[v]
        print(f"    {d['label']}: {da} "
              f"(rank 1 em {n} de {len(SCENARIOS)} cenarios)")

# Front Pareto
print("\nFront Pareto -- tamanho por versao (SC1)")
print("-" * 40)
for v, d in VERSIONS.items():
    sizes = [len(d['pareto_front'][sc]) for sc in SCENARIOS]
    invariante = len(set(
        tuple(sorted(str(c) for c in d['pareto_front'][sc]))
        for sc in SCENARIOS
    )) == 1
    print(f"  {d['label']}: SC1-SC4 = {sizes} | "
          f"invariante entre cenarios: {'Sim' if invariante else 'Nao'}")

# HMMD
print("\nHMMD -- front por versao")
print("-" * 25)
for v, d in VERSIONS.items():
    sizes = [len(d['hmmd'][sc]) for sc in SCENARIOS]
    invariante = len(set(
        tuple(sorted(str(c) for c in d['hmmd'][sc]))
        for sc in SCENARIOS
    )) == 1
    configs = set(
        c for sc in SCENARIOS for c in d['hmmd'][sc]
    )
    print(f"  {d['label']}: SC1-SC4 = {sizes} | "
          f"invariante: {'Sim' if invariante else 'Nao'} | "
          f"configs unicas: {len(configs)}")
```

```python
# Scores do Vetor A por versao
# V1: E3.2 do registro v10.0 -- escala 1-5, julgamento preliminar
# V2: E6.1 do registro v10.0 -- escala 1-5, consolidado 4 LLMs
# V3: notebook principal -- escala 1-4, consolidado 4 LLMs pos-revisao
# Ordem colunas: C1, C2, C3, C4, C5

VETOR_A = {
    'V1': {
        'D1':  [5, 4, 4, 5, 3],
        'D2':  [5, 5, 5, 4, 4],
        'D3':  [2, 2, 3, 3, 5],
        'D4':  [3, 2, 1, 3, 5],
        'I1':  [5, 5, 5, 5, 3],
        'I2':  [4, 4, 4, 5, 4],
        'I3':  [3, 2, 2, 3, 5],
        'I4':  [3, 3, 3, 3, 4],
        'ID1': [5, 5, 5, 4, 3],
        'ID2': [4, 4, 4, 5, 4],
        'ID3': [2, 2, 3, 3, 5],
        'R1':  [5, 5, 5, 5, 3],
        'R2':  [4, 3, 3, 4, 5],
        'R3':  [3, 2, 2, 2, 4],
    },
    'V2': {
        'D1':  [5, 3, 4, 5, 4],
        'D2':  [4, 4, 5, 5, 3],
        'D3':  [1, 1, 2, 3, 4],
        'D4':  [4, 3, 2, 2, 3],
        'I1':  [5, 3, 4, 5, 4],
        'I2':  [5, 3, 4, 5, 3],
        'I3':  [4, 2, 2, 3, 4],
        'I4':  [3, 3, 4, 4, 4],
        'ID1': [5, 5, 5, 5, 3],
        'ID2': [3, 3, 4, 5, 4],
        'ID3': [1, 2, 2, 3, 3],
        'R1':  [5, 5, 5, 5, 3],
        'R2':  [4, 3, 3, 4, 4],
        'R3':  [5, 4, 5, 4, 4],
    },
    'V3': {
        'D1':  [4, 3, 4, 4, 3],
        'D2':  [4, 4, 4, 4, 3],
        'D3':  [1, 1, 2, 2, 3],
        'D4':  [3, 2, 2, 2, 2],
        'I1':  [4, 3, 3, 4, 3],
        'I2':  [4, 4, 4, 4, 3],
        'I3':  [3, 2, 2, 3, 3],
        'I4':  [3, 2, 2, 3, 3],
        'ID1': [4, 4, 4, 4, 2],
        'ID2': [3, 2, 3, 3, 3],
        'ID3': [1, 1, 2, 2, 3],
        'R1':  [4, 4, 4, 4, 3],
        'R2':  [3, 3, 3, 3, 4],
        'R3':  [4, 4, 4, 4, 3],
    },
}

CRITERIA = ['C1', 'C2', 'C3', 'C4', 'C5']
MODULES = {
    'D':  ['D1', 'D2', 'D3', 'D4'],
    'I':  ['I1', 'I2', 'I3', 'I4'],
    'ID': ['ID1', 'ID2', 'ID3'],
    'R':  ['R1', 'R2', 'R3'],
}
SCENARIOS = {
    'SC1': {'C1': .20, 'C2': .20, 'C3': .20, 'C4': .20, 'C5': .20},
    'SC2': {'C1': .35, 'C2': .10, 'C3': .10, 'C4': .35, 'C5': .10},
    'SC3': {'C1': .10, 'C2': .20, 'C3': .20, 'C4': .10, 'C5': .40},
    'SC4': {'C1': .10, 'C2': .35, 'C3': .35, 'C4': .10, 'C5': .10},
}

def rank_das(version, module, scenario):
    """
    Rankeia as DAs de um modulo pelo score ponderado de um cenario.
    Retorna lista de (DA, score, rank) ordenada por rank.
    """
    w = SCENARIOS[scenario]
    das = MODULES[module]
    scores = {
        da: sum(w[c] * VETOR_A[version][da][i]
                for i, c in enumerate(CRITERIA))
        for da in das
    }
    sorted_scores = sorted(set(scores.values()), reverse=True)
    rank_map = {s: i+1 for i, s in enumerate(sorted_scores)}
    return sorted(
        [(da, round(scores[da], 4), rank_map[scores[da]]) for da in das],
        key=lambda x: x[2]
    )

print("Scores carregados:")
for v in VETOR_A:
    n = sum(len(das) for das in MODULES.values())
    print(f"  {v}: {n} DAs, escala {VERSIONS[v]['vetor_a_scale']}")
```

### Ranking por modulo e cenario

```python
# Ranking por modulo e cenario -- comparacao entre versoes
VERSIONS_LIST = ['V1', 'V2', 'V3']

for mod, das in MODULES.items():
    print(f"\n{'='*70}")
    print(f"MODULO {mod}")
    print(f"{'='*70}")

    for sc, w in SCENARIOS.items():
        print(f"\n{sc}")
        print(f"{'DA':<6} {'V1 score':>10} {'V1 rank':>8} "
              f"{'V2 score':>10} {'V2 rank':>8} "
              f"{'V3 score':>10} {'V3 rank':>8}")
        print("-" * 65)

        # Calcular scores e ranks por versao
        all_ranks = {da: {} for da in das}
        for v in VERSIONS_LIST:
            ranked = rank_das(v, mod, sc)
            for da, score, rank in ranked:
                all_ranks[da][v] = (score, rank)

        # Ordenar por rank medio
        avg_rank = {
            da: sum(all_ranks[da][v][1] for v in VERSIONS_LIST) / len(VERSIONS_LIST)
            for da in das
        }
        for da in sorted(das, key=lambda x: avg_rank[x]):
            row = f"{da:<6}"
            for v in VERSIONS_LIST:
                score, rank = all_ranks[da][v]
                row += f" {score:>10.4f} {rank:>8}"
            print(row)
```

```python
# Variacao de rank entre versoes por DA e cenario
print("Variacao de rank entre versoes (max - min rank por DA e cenario)")
print("=" * 70)

for mod, das in MODULES.items():
    print(f"\nModulo {mod}")
    print(f"{'DA':<6} {'SC1':>6} {'SC2':>6} {'SC3':>6} {'SC4':>6} {'Media':>8} {'Max':>6}")
    print("-" * 45)

    for da in das:
        variacoes = []
        for sc in SCENARIOS:
            ranks = [rank_das(v, mod, sc) for v in VERSIONS_LIST]
            rank_por_versao = {
                v: next(r for d, s, r in rank_das(v, mod, sc) if d == da)
                for v in VERSIONS_LIST
            }
            variacao = max(rank_por_versao.values()) - min(rank_por_versao.values())
            variacoes.append(variacao)

        media = sum(variacoes) / len(variacoes)
        maximo = max(variacoes)
        print(f"{da:<6} {variacoes[0]:>6} {variacoes[1]:>6} "
              f"{variacoes[2]:>6} {variacoes[3]:>6} "
              f"{media:>8.2f} {maximo:>6}")
```

```python
# Sintese da variacao de rank
print("Sintese -- DA mais estavel e mais instavel por modulo")
print("=" * 70)

for mod, das in MODULES.items():
    max_variacao_possivel = len(das) - 1
    print(f"\nModulo {mod} (variacao maxima possivel: {max_variacao_possivel})")
    print("-" * 50)

    # Calcular variacao media por DA
    var_media = {}
    for da in das:
        variacoes = []
        for sc in SCENARIOS:
            rank_por_versao = {
                v: next(r for d, s, r in rank_das(v, mod, sc) if d == da)
                for v in VERSIONS_LIST
            }
            variacoes.append(max(rank_por_versao.values()) - min(rank_por_versao.values()))
        var_media[da] = sum(variacoes) / len(variacoes)

    mais_estavel = min(var_media, key=var_media.get)
    mais_instavel = max(var_media, key=var_media.get)

    # DA dominante por versao (rank 1 mais frequente)
    dominante_por_versao = {}
    for v in VERSIONS_LIST:
        contagem = {}
        for sc in SCENARIOS:
            da_rank1 = next(d for d, s, r in rank_das(v, mod, sc) if r == 1)
            contagem[da_rank1] = contagem.get(da_rank1, 0) + 1
        dominante_por_versao[v] = max(contagem, key=contagem.get)

    print(f"  DA mais estavel:   {mais_estavel} "
          f"(variacao media: {var_media[mais_estavel]:.2f})")
    print(f"  DA mais instavel:  {mais_instavel} "
          f"(variacao media: {var_media[mais_instavel]:.2f})")
    print(f"  DA dominante (rank 1 mais frequente):")
    for v in VERSIONS_LIST:
        dom = dominante_por_versao[v]
        n = sum(1 for sc in SCENARIOS
                if next(r for d, s, r in rank_das(v, mod, sc) if d == dom) == 1)
        print(f"    {v}: {dom} ({n} de {len(SCENARIOS)} cenarios)")
```

```python
# Mudancas de ranking -- foco na DA
print("Mudancas de ranking por DA")
print("=" * 70)
comparacoes = [('V1', 'V2'), ('V2', 'V3'), ('V1', 'V3')]

for mod, das in MODULES.items():
    print(f"\nModulo {mod}")
    print("=" * 50)

    for da in das:
        mudancas = []
        for v_from, v_to in comparacoes:
            for sc in SCENARIOS:
                r_from = next(r for d, s, r in rank_das(v_from, mod, sc) if d == da)
                r_to = next(r for d, s, r in rank_das(v_to, mod, sc) if d == da)
                delta = r_to - r_from
                if delta != 0:
                    mudancas.append((v_from, v_to, sc, r_from, r_to, delta))

        if mudancas:
            print(f"\n  {da}")
            print(f"  {'Transicao':<12} {'Cenario':<8} "
                  f"{'Antes':>7} {'Depois':>8} {'Delta':>7}")
            print("  " + "-" * 46)
            for v_from, v_to, sc, r_from, r_to, delta in mudancas:
                sinal = '+' if delta > 0 else ''
                print(f"  {v_from}->{v_to:<8} {sc:<8} "
                      f"{r_from:>7} {r_to:>8} {sinal:>6}{delta}")
        else:
            print(f"\n  {da}: sem mudancas em nenhum cenario ou transicao")
```

```python
# Criterio responsavel pela mudanca de rank
print("Criterio responsavel pela mudanca de rank")
print("=" * 70)

for mod, das in MODULES.items():
    print(f"\nModulo {mod}")
    print("=" * 50)

    for v_from, v_to in [('V1', 'V2'), ('V2', 'V3'), ('V1', 'V3')]:
        inversoes = []
        for sc, w in SCENARIOS.items():
            ranks_from = {d: r for d, s, r in rank_das(v_from, mod, sc)}
            ranks_to   = {d: r for d, s, r in rank_das(v_to,   mod, sc)}

            # Identificar pares que trocaram de posicao relativa
            for i, da1 in enumerate(das):
                for da2 in das[i+1:]:
                    rel_from = ranks_from[da1] - ranks_from[da2]
                    rel_to   = ranks_to[da1]   - ranks_to[da2]
                    if (rel_from > 0 and rel_to < 0) or \
                       (rel_from < 0 and rel_to > 0):
                        # Calcular diferenca de score por criterio
                        deltas_c = {}
                        for j, c in enumerate(CRITERIA):
                            s1_from = VETOR_A[v_from][da1][j]
                            s2_from = VETOR_A[v_from][da2][j]
                            s1_to   = VETOR_A[v_to][da1][j]
                            s2_to   = VETOR_A[v_to][da2][j]
                            # Diferenca ponderada entre da1 e da2
                            diff_from = w[c] * (s1_from - s2_from)
                            diff_to   = w[c] * (s1_to   - s2_to)
                            # Criterio que mais contribuiu para a inversao
                            deltas_c[c] = abs(diff_to - diff_from)
                        criterio_resp = max(deltas_c, key=deltas_c.get)
                        inversoes.append((sc, da1, da2,
                                          ranks_from[da1], ranks_from[da2],
                                          ranks_to[da1],   ranks_to[da2],
                                          criterio_resp, deltas_c[criterio_resp]))

        if inversoes:
            print(f"\n  {v_from} -> {v_to}")
            print(f"  {'Cenario':<6} {'DA1':<5} {'DA2':<5} "
                  f"{'Rank antes':>12} {'Rank depois':>12} "
                  f"{'Criterio':>10} {'Delta pond':>11}")
            print("  " + "-" * 65)
            for sc, da1, da2, rf1, rf2, rt1, rt2, crit, delta in inversoes:
                print(f"  {sc:<6} {da1:<5} {da2:<5} "
                      f"{rf1}/{rf2}{'':>8} {rt1}/{rt2}{'':>8} "
                      f"{crit:>10} {delta:>11.4f}")
        else:
            print(f"\n  {v_from} -> {v_to}: sem inversoes")
```
