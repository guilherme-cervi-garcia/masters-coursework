## Solution Space Exploration

### Configuration Generation and Filtering

````python
def generate_configurations():
    """Generate all 144 configurations from D × I × ID × R product."""
    configs = []
    for d, i, id_, r in itertools.product(
        MODULE_VALUES['D'],
        MODULE_VALUES['I'],
        MODULE_VALUES['ID'],
        MODULE_VALUES['R']
    ):
        configs.append({'D': d, 'I': i, 'ID': id_, 'R': r})
    return configs

def compute_w(config):
    """Compute w(S) = minimum pairwise CCM compatibility.
    Per Levin (2012, Section 3.7): N(S) = (w(S); e(S)).
    """
    das = [config['D'], config['I'], config['ID'], config['R']]
    scores = []
    for da1, da2 in itertools.combinations(das, 2):
        score = get_ccm(da1, da2)
        if score is None:
            raise ValueError(f"Pair not found in CCM: ({da1}, {da2})")
        scores.append(score)
    return min(scores)

# Generate and score all configurations
all_configs = generate_configurations()
for cfg in all_configs:
    cfg['w'] = compute_w(cfg)

# Filter admissible
admissible = [c for c in all_configs if c['w'] >= 2]
for i, cfg in enumerate(admissible, 1):
    cfg['id'] = f"S{i:02d}"

print(f"Configurations generated: {len(all_configs)}")
print(f"Admissible (w ≥ 2): {len(admissible)}")
print(f"  w(S) = 3: {sum(1 for c in admissible if c['w'] == 3)}")
print(f"  w(S) = 2: {sum(1 for c in admissible if c['w'] == 2)}")
````

## Selection Methods Implementation

Four methods applied to 144 admissible configurations, across 4 scenarios.

````python
# Core selection method implementations

def scoring_linear(admissible, scenario, vetor_a):
    """Weighted linear scoring: rank by Q(S) descending."""
    w_mod = scenario['module_weights']
    w_crit = scenario['criteria_weights']
    results = []
    for cfg in admissible:
        q = 0.0
        for mod in MODULES:
            da = cfg[mod]
            da_score = sum(
                w_crit[c] * vetor_a.loc[da, c] for c in CRITERIA
            )
            q += w_mod[mod] * da_score
        results.append({'id': cfg['id'], 'Q': round(q, 4), 'config': cfg})
    results.sort(key=lambda x: x['Q'], reverse=True)
    for rank, r in enumerate(results, 1):
        r['rank'] = rank
    return results

def ideal_point(admissible, scenario, vetor_a):
    """Euclidean distance to ideal point: rank by ρ ascending."""
    w_mod = scenario['module_weights']
    w_crit = scenario['criteria_weights']
    
    # Ideal point per module
    ideal = {}
    for mod in MODULES:
        best_da = max(
            MODULE_VALUES[mod],
            key=lambda da: sum(w_crit[c] * vetor_a.loc[da, c] for c in CRITERIA)
        )
        ideal[mod] = {c: vetor_a.loc[best_da, c] for c in CRITERIA}
    
    results = []
    for cfg in admissible:
        rho_sq = 0.0
        for mod in MODULES:
            da = cfg[mod]
            for c in CRITERIA:
                diff = ideal[mod][c] - vetor_a.loc[da, c]
                rho_sq += w_mod[mod] * w_crit[c] * (diff ** 2)
        rho = round(rho_sq ** 0.5, 4)
        results.append({'id': cfg['id'], 'rho': rho, 'config': cfg})
    results.sort(key=lambda x: x['rho'])
    for rank, r in enumerate(results, 1):
        r['rank'] = rank
    return results

def compute_criteria_scores(admissible, scenario, vetor_a):
    """Compute aggregated criterion scores per config."""
    w_mod = scenario['module_weights']
    scores = {}
    for cfg in admissible:
        scores[cfg['id']] = {
            c: sum(w_mod[mod] * vetor_a.loc[cfg[mod], c] for mod in MODULES)
            for c in CRITERIA
        }
    return scores

def pareto_based_ma(admissible, scenario, vetor_a):
    """Identify Pareto-efficient configurations."""
    crit_scores = compute_criteria_scores(admissible, scenario, vetor_a)
    ids = [cfg['id'] for cfg in admissible]
    dominated = set()
    
    for s1 in ids:
        for s2 in ids:
            if s1 == s2:
                continue
            geq_all = all(crit_scores[s2][c] >= crit_scores[s1][c] for c in CRITERIA)
            gt_one  = any(crit_scores[s2][c] >  crit_scores[s1][c] for c in CRITERIA)
            if geq_all and gt_one:
                dominated.add(s1)
                break
    
    pareto_front = [cfg for cfg in admissible if cfg['id'] not in dominated]
    return pareto_front, crit_scores

def compute_priorities(scenario, vetor_a):
    """Compute ordinal priorities per module."""
    w_crit = scenario['criteria_weights']
    priorities = {}
    for mod in MODULES:
        das = MODULE_VALUES[mod]
        scores = {
            da: sum(w_crit[c] * vetor_a.loc[da, c] for c in CRITERIA)
            for da in das
        }
        sorted_scores = sorted(set(scores.values()), reverse=True)
        rank_map = {s: i+1 for i, s in enumerate(sorted_scores)}
        priorities[mod] = {da: rank_map[scores[da]] for da in das}
    return priorities

def compute_e(config, priorities, k=4):
    """Compute e(S) multiset."""
    e = [0] * k
    for mod in MODULES:
        da = config[mod]
        rank = priorities[mod][da]
        if rank <= k:
            e[rank - 1] += 1
    return tuple(e)

def hmmd_dominates(n1, n2):
    """Lexicographic dominance check on N = (w; e)."""
    w1, e1 = n1
    w2, e2 = n2
    w_geq = w1 >= w2
    w_gt  = w1 > w2
    e_geq = e1 >= e2
    e_gt  = e1 > e2
    return (w_geq and e_geq) and (w_gt or e_gt)

def hmmd(admissible, scenario, vetor_a, k=4):
    """Identify HMMD-non-dominated configurations."""
    priorities = compute_priorities(scenario, vetor_a)
    
    ns = {}
    for cfg in admissible:
        w = cfg['w']
        e = compute_e(cfg, priorities, k)
        ns[cfg['id']] = (w, e)
    
    dominated = set()
    ids = [cfg['id'] for cfg in admissible]
    for s1 in ids:
        for s2 in ids:
            if s1 == s2:
                continue
            if hmmd_dominates(ns[s2], ns[s1]):
                dominated.add(s1)
                break
    
    pareto_hmmd = [cfg for cfg in admissible if cfg['id'] not in dominated]
    return pareto_hmmd, ns, priorities

# Execute all methods for all scenarios
scoring_results = {}
ideal_results = {}
pareto_results = {}
hmmd_results = {}

for sc_id, sc in SCENARIOS.items():
    scoring_results[sc_id] = scoring_linear(admissible, sc, vetor_a)
    ideal_results[sc_id] = ideal_point(admissible, sc, vetor_a)
    front, crit_scores = pareto_based_ma(admissible, sc, vetor_a)
    pareto_results[sc_id] = {'front': front, 'crit_scores': crit_scores}
    front_hmmd, ns, priorities = hmmd(admissible, sc, vetor_a)
    hmmd_results[sc_id] = {'front': front_hmmd, 'ns': ns, 'priorities': priorities}

print("\n✓ All 4 selection methods executed across 4 scenarios.")
print("\nPareto front sizes (V3 post-recalibration):")
for sc_id, sc in SCENARIOS.items():
    p_size = len(pareto_results[sc_id]['front'])
    h_size = len(hmmd_results[sc_id]['front'])
    print(f"  {sc_id}: Pareto={p_size}, HMMD={h_size}")
````

## Solution Space Exploration

### Configuration Generation

All 144 configurations from the Cartesian product D × I × ID × R.

````python
def generate_configurations():
    """Generate all 144 configurations from D × I × ID × R product."""
    configs = []
    for d, i, id_, r in itertools.product(
        MODULE_VALUES['D'],
        MODULE_VALUES['I'],
        MODULE_VALUES['ID'],
        MODULE_VALUES['R']
    ):
        configs.append({'D': d, 'I': i, 'ID': id_, 'R': r})
    return configs

def compute_w(config):
    """Compute w(S) = minimum pairwise CCM compatibility.
    Per Levin (2012, Section 3.7): N(S) = (w(S); e(S)).
    6 pairs correspond to blocks B1-B6 of the CCM.
    """
    das = [config['D'], config['I'], config['ID'], config['R']]
    scores = []
    for da1, da2 in itertools.combinations(das, 2):
        score = get_ccm(da1, da2)
        if score is None:
            raise ValueError(f"Pair not found in CCM: ({da1}, {da2})")
        scores.append(score)
    return min(scores)

# Generate and score all configurations
all_configs = generate_configurations()
for cfg in all_configs:
    cfg['w'] = compute_w(cfg)

print(f"Configurations generated: {len(all_configs)}")
print(f"  w(S) = 3: {sum(1 for c in all_configs if c['w'] == 3)}")
print(f"  w(S) = 2: {sum(1 for c in all_configs if c['w'] == 2)}")
print(f"  w(S) = 1: {sum(1 for c in all_configs if c['w'] == 1)}")
print(f"  w(S) = 0: {sum(1 for c in all_configs if c['w'] == 0)}")
````

### Admissible Space Filtering

Criterion: w(S) ≥ 2 (threshold for moderate compatibility or better).

The CCM contains no zero-score pairs—all configurations have at least conditional compatibility.

````python
def filter_admissible(configs, threshold=2):
    """Filter to admissible configurations with w(S) >= threshold."""
    admissible = [c for c in configs if c['w'] >= threshold]
    for i, cfg in enumerate(admissible, 1):
        cfg['id'] = f"S{i:02d}"
    return admissible

for cfg in all_configs:
    cfg.pop('id', None)

admissible = filter_admissible(all_configs, threshold=2)

print(f"Admissible space: {len(admissible)} of {len(all_configs)} configurations")
print(f"Reduction: {100*(1 - len(admissible)/len(all_configs)):.1f}%")
print(f"  w(S) = 3: {sum(1 for c in admissible if c['w'] == 3)} configurations")
print(f"  w(S) = 2: {sum(1 for c in admissible if c['w'] == 2)} configurations")
````

## Selection Methods

Four methods applied to the admissible space (144 configurations).

### Weighted Linear Scoring

$$Q(S) = \sum_{m \in \text{MODULES}} w_m \cdot \sum_{c \in \text{CRITERIA}} w_c \cdot \text{score}(da_m, c)$$

````python
def scoring_linear(admissible, scenario, vetor_a):
    """Compute Q(S) for all configurations; rank descending."""
    w_mod = scenario['module_weights']
    w_crit = scenario['criteria_weights']
    results = []

    for cfg in admissible:
        q = 0.0
        for mod in MODULES:
            da = cfg[mod]
            da_score = sum(
                w_crit[c] * vetor_a.loc[da, c]
                for c in CRITERIA
            )
            q += w_mod[mod] * da_score
        results.append({'id': cfg['id'], 'Q': round(q, 4)})

    results.sort(key=lambda x: x['Q'], reverse=True)
    for rank, r in enumerate(results, 1):
        r['rank'] = rank
    return results

# Execute for all scenarios
print("Weighted Linear Scoring — Top 10 per scenario")
print("=" * 70)

scoring_results = {}
for sc_id, sc in SCENARIOS.items():
    res = scoring_linear(admissible, sc, vetor_a)
    scoring_results[sc_id] = res
    print(f"\n{sc_id} — {sc['name']}")
    print(f"{'Rank':<6} {'ID':<6} {'Q':>8}  {'D':<6} {'I':<6} {'ID':<6} {'R':<6}")
    print("-" * 65)
    cfg_map = {c['id']: c for c in admissible}
    for r in res[:10]:
        cfg = cfg_map[r['id']]
        print(f"{r['rank']:<6} {r['id']:<6} {r['Q']:>8.4f}  "
              f"{cfg['D']:<6} {cfg['I']:<6} {cfg['ID']:<6} {cfg['R']:<6}")
````

### Ideal Point Method

$$\rho(S, S^o) = \sqrt{\sum_{m} w_m \cdot \sum_{c} w_c \cdot (\text{score}^o_c - \text{score}(da_m, c))^2}$$

````python
def ideal_point(admissible, scenario, vetor_a):
    """Compute Euclidean distance to ideal point; rank ascending."""
    w_mod = scenario['module_weights']
    w_crit = scenario['criteria_weights']

    # Ideal point: best DA per module (highest weighted score)
    ideal = {}
    for mod in MODULES:
        best_da = max(
            MODULE_VALUES[mod],
            key=lambda da: sum(w_crit[c] * vetor_a.loc[da, c] for c in CRITERIA)
        )
        ideal[mod] = {c: vetor_a.loc[best_da, c] for c in CRITERIA}

    results = []
    for cfg in admissible:
        rho_sq = 0.0
        for mod in MODULES:
            da = cfg[mod]
            for c in CRITERIA:
                diff = ideal[mod][c] - vetor_a.loc[da, c]
                rho_sq += w_mod[mod] * w_crit[c] * (diff ** 2)
        rho = round(rho_sq ** 0.5, 4)
        results.append({'id': cfg['id'], 'rho': rho})

    results.sort(key=lambda x: x['rho'])
    for rank, r in enumerate(results, 1):
        r['rank'] = rank
    return results

# Execute for all scenarios
print("Ideal Point Method — Top 10 per scenario")
print("=" * 70)

ideal_results = {}
for sc_id, sc in SCENARIOS.items():
    res = ideal_point(admissible, sc, vetor_a)
    ideal_results[sc_id] = res
    print(f"\n{sc_id} — {sc['name']}")
    print(f"{'Rank':<6} {'ID':<6} {'rho':>9}  {'D':<6} {'I':<6} {'ID':<6} {'R':<6}")
    print("-" * 65)
    cfg_map = {c['id']: c for c in admissible}
    for r in res[:10]:
        cfg = cfg_map[r['id']]
        print(f"{r['rank']:<6} {r['id']:<6} {r['rho']:>9.4f}  "
              f"{cfg['D']:<6} {cfg['I']:<6} {cfg['ID']:<6} {cfg['R']:<6}")
````

### Pareto-Based Morphological Analysis

$$f_c(S) = \sum_{m \in \text{MODULES}} w_m \cdot \text{score}(da_m, c)$$

Configuration S is Pareto-efficient if no S' dominates it across all criteria.

````python
def compute_criteria_scores(admissible, scenario, vetor_a):
    """Compute aggregated criterion scores per configuration."""
    w_mod = scenario['module_weights']
    scores = {}
    for cfg in admissible:
        scores[cfg['id']] = {
            c: sum(w_mod[mod] * vetor_a.loc[cfg[mod], c] for mod in MODULES)
            for c in CRITERIA
        }
    return scores

def pareto_based_ma(admissible, scenario, vetor_a):
    """Identify Pareto-efficient configurations."""
    crit_scores = compute_criteria_scores(admissible, scenario, vetor_a)
    ids = [cfg['id'] for cfg in admissible]
    dominated = set()

    for s1 in ids:
        for s2 in ids:
            if s1 == s2:
                continue
            # Check if s1 is dominated by s2
            geq_all = all(crit_scores[s2][c] >= crit_scores[s1][c] for c in CRITERIA)
            gt_one  = any(crit_scores[s2][c] >  crit_scores[s1][c] for c in CRITERIA)
            if geq_all and gt_one:
                dominated.add(s1)
                break

    pareto_front = [cfg for cfg in admissible if cfg['id'] not in dominated]
    return pareto_front, crit_scores

# Execute for all scenarios
print("Pareto-Based MA — Pareto Front per scenario")
print("=" * 90)

pareto_results = {}
for sc_id, sc in SCENARIOS.items():
    front, crit_scores = pareto_based_ma(admissible, sc, vetor_a)
    pareto_results[sc_id] = {'front': front, 'crit_scores': crit_scores}

    print(f"\n{sc_id} — {sc['name']}")
    print(f"Pareto front: {len(front)} of {len(admissible)} configurations")
    print(f"{'ID':<6} {'D':<6} {'I':<6} {'ID':<6} {'R':<6} "
          f"{'f_C1':>7} {'f_C2':>7} {'f_C3':>7} {'f_C4':>7} {'f_C5':>7}")
    print("-" * 80)
    for cfg in sorted(front, key=lambda x: x['id']):
        cs = crit_scores[cfg['id']]
        print(f"{cfg['id']:<6} {cfg['D']:<6} {cfg['I']:<6} "
              f"{cfg['ID']:<6} {cfg['R']:<6} "
              f"{cs['C1']:>7.3f} {cs['C2']:>7.3f} {cs['C3']:>7.3f} "
              f"{cs['C4']:>7.3f} {cs['C5']:>7.3f}")
````

### Hierarchical Morphological Multicriteria Design (HMMD)

$$N(S) = (w(S); e(S))$$

Where e(S) is a multiset counting DAs by priority rank.

````python
def compute_priorities(scenario, vetor_a):
    """Compute ordinal priorities per module per scenario."""
    w_crit = scenario['criteria_weights']
    priorities = {}

    for mod in MODULES:
        das = MODULE_VALUES[mod]
        scores = {
            da: sum(w_crit[c] * vetor_a.loc[da, c] for c in CRITERIA)
            for da in das
        }
        # Dense ranking (ties get same rank)
        sorted_scores = sorted(set(scores.values()), reverse=True)
        rank_map = {s: i+1 for i, s in enumerate(sorted_scores)}
        priorities[mod] = {da: rank_map[scores[da]] for da in das}

    return priorities

def compute_e(config, priorities, k=4):
    """Compute e(S) multiset: counts of DAs per rank."""
    e = [0] * k
    for mod in MODULES:
        da = config[mod]
        rank = priorities[mod][da]
        if rank <= k:
            e[rank - 1] += 1
    return tuple(e)

def hmmd_dominates(n1, n2):
    """Check if N1 lexicographically dominates N2."""
    w1, e1 = n1
    w2, e2 = n2
    w_geq = w1 >= w2
    w_gt  = w1 > w2
    e_geq = e1 >= e2
    e_gt  = e1 > e2
    return (w_geq and e_geq) and (w_gt or e_gt)

def hmmd(admissible, scenario, vetor_a, k=4):
    """Identify HMMD-non-dominated configurations."""
    priorities = compute_priorities(scenario, vetor_a)

    ns = {}
    for cfg in admissible:
        w = cfg['w']
        e = compute_e(cfg, priorities, k)
        ns[cfg['id']] = (w, e)

    dominated = set()
    ids = [cfg['id'] for cfg in admissible]
    for s1 in ids:
        for s2 in ids:
            if s1 == s2:
                continue
            if hmmd_dominates(ns[s2], ns[s1]):
                dominated.add(s1)
                break

    pareto_hmmd = [cfg for cfg in admissible if cfg['id'] not in dominated]
    return pareto_hmmd, ns, priorities

# Execute for all scenarios
print("HMMD — Pareto Front per scenario")
print("=" * 90)

hmmd_results = {}
for sc_id, sc in SCENARIOS.items():
    front, ns, priorities = hmmd(admissible, sc, vetor_a)
    hmmd_results[sc_id] = {'front': front, 'ns': ns, 'priorities': priorities}

    print(f"\n{sc_id} — {sc['name']}")
    print(f"HMMD front: {len(front)} of {len(admissible)} configurations")
    print(f"{'ID':<6} {'D':<6} {'I':<6} {'ID':<6} {'R':<6} "
          f"{'w(S)':>6} {'e(S)':<20}")
    print("-" * 80)
    for cfg in sorted(front, key=lambda x: x['id']):
        n = ns[cfg['id']]
        print(f"{cfg['id']:<6} {cfg['D']:<6} {cfg['I']:<6} "
              f"{cfg['ID']:<6} {cfg['R']:<6} "
              f"{n[0]:>6} {str(n[1]):<20}")
````

## Comparative Analysis

Spearman correlation, method divergences, and cross-scenario robustness.

````python
# 5.1 Spearman correlation — Scoring vs Ideal Point
print("Spearman Correlation — Scoring vs Ideal Point")
print("-" * 50)
print(f"{'Scenario':<10} {'rho':>10} {'p':>10}")
print("-" * 30)

spearman_results = {}
for sc_id in SCENARIOS:
    s_res = scoring_results[sc_id]
    i_res = ideal_results[sc_id]

    ids = [r['id'] for r in s_res]
    s_ranks = [r['rank'] for r in s_res]
    i_rank_map = {r['id']: r['rank'] for r in i_res}
    i_ranks = [i_rank_map[sid] for sid in ids]

    rho, p = spearmanr(s_ranks, i_ranks)
    spearman_results[sc_id] = rho
    print(f"{sc_id:<10} {rho:>10.4f} {p:>10.4f}")

print(f"\nSpearman minimum (across scenarios): {min(spearman_results.values()):.4f}")

# 5.2 Robustness analysis
print("\nRobustness Analysis — Top 5 configurations by mean rank")
print("-" * 80)

cfg_map = {c['id']: c for c in admissible}
rank_stats = {}
sc_ids = list(SCENARIOS.keys())

for sid in [c['id'] for c in admissible]:
    ranks = [r['rank'] for sc in sc_ids
             for r in scoring_results[sc] if r['id'] == sid]
    rank_stats[sid] = {
        'mean': sum(ranks) / len(ranks),
        'max': max(ranks)
    }

sorted_by_robustness = sorted(
    rank_stats.items(),
    key=lambda x: (x[1]['mean'], x[1]['max'])
)

print(f"{'ID':<6} {'D':<6} {'I':<6} {'ID':<6} {'R':<6} {'Mean Rank':>10} {'Max Rank':>10}")
print("-" * 70)
for sid, stats in sorted_by_robustness[:5]:
    cfg = cfg_map[sid]
    print(f"{sid:<6} {cfg['D']:<6} {cfg['I']:<6} {cfg['ID']:<6} {cfg['R']:<6} "
          f"{stats['mean']:>10.2f} {stats['max']:>10}")
````
